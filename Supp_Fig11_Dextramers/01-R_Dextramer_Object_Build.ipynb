# Build

```r
build_so_func <- function(h5_path){
    h5_list <- h5dump(h5_path)
    rna_mtx <- h5_list_convert_to_dgCMatrix(h5_list, target = 'matrix', feature_names = 'name')
    adt_mtx <- h5_list_convert_to_dgCMatrix(h5_list, target = 'ADT', feature_names = 'name')
    cell_meta <- read_h5_cell_meta(h5_path)

    so <- CreateSeuratObject(counts = rna_mtx$matrix_dgCMatrix, meta.data = cell_meta)
    so[['ADT']] <- CreateAssayObject(counts = adt_mtx$ADT_dgCMatrix)
    return(so)
}
```

```r
so_list <- lapply(h5_trim, build_so_func)
```

```r
so <- Reduce(merge, so_list)
```

## Add TCR

```r
add_tcr_func_sample <- function(seurat_object, pool){
    # build barcode dataframe
    barcodes_df <- select(seurat_object@meta.data, classic_barcodes, barcodes)

    # extract sample ID
    sample_id <- unique(seurat_object$sample_id)

    # subset T cells
    t_cells <- subset(seurat_object, AIFI_PBMC.Flex_L1 == 'T cell')
    non_t_cells <- subset(seurat_object, AIFI_PBMC.Flex_L1 != 'T cell')
    
    # load tcr
    tcr <- read.csv(paste0('tcr/',pool,'-clones/','EXP-1589-DEX-',pool,'.',sample_id,'.alignments.clones.tsv'),sep='\t')

    # replace barcodes with barcodes from pipeline
    tcr_updated <- tcr %>%
        left_join(barcodes_df, by = c("tagValueCELL" = "classic_barcodes")) %>%
        mutate(tagValueCELL = ifelse(!is.na(barcodes), barcodes, tagValueCELL)) %>%
        select(-barcodes)
    # trim barcodes for any non-TCR clones
    tcr_updated <- tcr_updated[!(tcr_updated$topChains %in% c("IGH", "IGK", "IGL")), ]
    # load via scRep
    tcr_contigs <- loadContigs(tcr_updated, format = 'MiXCR')
    combined.TCR <- combineTCR(tcr_contigs,
                               removeNA = FALSE, 
                               removeMulti = FALSE, 
                               filterMulti = TRUE)
    t_cells <- combineExpression(combined.TCR, 
                                 t_cells, 
                                 cloneCall="gene",
                                 proportion = TRUE)

    seurat_object <- merge(t_cells, non_t_cells)

    return(seurat_object)
}
```

```r
add_tcr_func_batch <- function(seurat_object,pool){
    # split object by sample
    print("Splitting Object")
    flush.console()
    so_split <- SplitObject(seurat_object, split.by = 'sample_id')
    # loop through and add tcr info
    print("Adding TCR")
    flush.console()
    so_split <- lapply(so_split, function(x){
        x <- add_tcr_func_sample(x, pool)
        return(x)
    })
    print("Merging Object")
    flush.console()
    seurat_object <- Reduce(merge, so_split)    
    return(seurat_object)
}
```

```r
dex_stained$classic_barcodes <- substr(dex_stained$original_barcodes, 1, nchar(dex_stained$original_barcodes)-8)
```

```r
pool_split <- SplitObject(dex_stained, split.by = 'pool_id')
```

```r
pool_split$`EXP-01589-PE` <- add_tcr_func_batch(seurat_object = pool_split$`EXP-01589-PE`, pool = 'PA')
```

```r
pool_split$`EXP-01589-PF` <- add_tcr_func_batch(seurat_object = pool_split$`EXP-01589-PF`, pool = 'PB')
```

```r
dex_tcr <- Reduce(merge, pool_split)
```
