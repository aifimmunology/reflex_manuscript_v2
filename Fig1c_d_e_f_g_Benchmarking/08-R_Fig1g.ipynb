# Clones Proportion

```r
filtered_metadata <- so@meta.data %>%
  filter(!is.na(CTaa)) %>%
  filter(!grepl("NA", CTaa))
```

```r
seurat_subset <- subset(so, cells = rownames(filtered_metadata))
```

```r
clone_scatter_df <- clonalScatter(seurat_subset, group.by = 'tech',exportTable = T,
              cloneCall ="aa", 
              x.axis = "REFLEX", 
              y.axis = "10x_5prime",
              dot.size = "total",
              graph = "proportion")
```

```r
target_clones <- c("CAGVGSTSGTYKYIF_CASSEANLLAVYNNEQFF",
                   "CARGAQKLVF_CASSQYRGAGASYEQYF",
                   "CALRGMNRDDKIIF_CASSISAQPQHF",
                   "CALKGLRGSNYKLTF_CASSDGAEQYF")
```

```r
clone_scatter_df %>%
    filter(Var1 %in% target_clones)
```

```r
clone_scatter_df <- clone_scatter_df %>%
  mutate(clone_color = if_else(Var1 %in% target_clones, Var1, "Other")) %>%
  arrange(desc(clone_color == "Other"))
```

```r
label_df <- clone_scatter_df %>%
  filter(Var1 %in% target_clones)
```

```r
options(repr.plot.width = 8, repr.plot.height = 8)
ggplot(clone_scatter_df %>% filter(REFLEX.fraction > 0.0001, `10x_5prime.fraction` > 0.0001), 
       aes(x = REFLEX.fraction, y = `10x_5prime.fraction`, color = clone_color, size = size)) + 
    geom_point(alpha = 0.8) + 
    geom_abline(slope = 1, intercept = 0, linetype = "dashed", color = "grey50") + 
    geom_text_repel(
        data = label_df,
        aes(label = Var1),
        size = 4,
        color = "black",
        nudge_x = 0.004, 
        nudge_y = -0.002,
        direction = "both",
        hjust = 0,
        vjust = 1,
        box.padding = 0.5, 
        point.padding = 0.3,
        show.legend = FALSE,
        min.segment.length = 0 
    ) + 
    scale_size_continuous(range = c(1, 15)) +
    scale_color_manual(values = c(
      "CAGVGSTSGTYKYIF_CASSEANLLAVYNNEQFF" = '#F8766D',
      "CARGAQKLVF_CASSQYRGAGASYEQYF" = '#C77CFF',
      "CALRGMNRDDKIIF_CASSISAQPQHF" = '#00BFC4',
      "CALKGLRGSNYKLTF_CASSDGAEQYF" = '#7CAE00', 
      "Other" = "lightgrey"
    )) +
    xlim(0, 0.025) + 
    ylim(0, 0.025) + 
    xlab('REFLEX Paired Clone Fraction') +
    ylab("10x 5' Paired Clone Fraction") +
    theme_classic(base_size = 20) + theme(legend.position = 'none')
```

# Clones Expression

```r
top_clones_so <- subset(so, CTaa %in% target_clones)
```

```r
top_clones_so_reflex <- subset(top_clones_so, tech == 'REFLEX')
top_clones_so_5prime <- subset(top_clones_so, tech == '10x_5prime')
```

```r
degs_top_clones <- FindAllMarkers(top_clones_so, group.by = 'CTaa')
```

```r
degs_top_clones %>%
    filter(avg_log2FC > 0) %>%
    group_by(cluster) %>%
    slice_min(order_by = p_val_adj, n=4, with_ties = FALSE) %>%
    ungroup()
```

```r
options(repr.plot.width = 16, repr.plot.height = 16)
p1 <- VlnPlot(top_clones_so_5prime, features = 'GZMK', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p2 <- VlnPlot(top_clones_so_reflex, features = 'GZMK', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p3 <- VlnPlot(top_clones_so_5prime, features = 'IKZF2', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p4 <- VlnPlot(top_clones_so_reflex, features = 'IKZF2', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p5 <- VlnPlot(top_clones_so_5prime, features = 'KLRC1', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p6 <- VlnPlot(top_clones_so_reflex, features = 'KLRC1', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p7 <- VlnPlot(top_clones_so_5prime, features = 'GNLY', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
p8 <- VlnPlot(top_clones_so_reflex, features = 'GNLY', group.by = 'CTaa') & theme(axis.text.x = element_blank(), legend.position='none')
clone_violins <- cowplot::plot_grid(p1,p2,p3,p4,p5,p6,p7,p8,ncol=2)
```
